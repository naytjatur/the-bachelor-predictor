```python
def first_impression_rose(df,col):
    df.loc[df[col] == 'R1','First Impression Rose?'] = True
    return df
```

```python
df = first_impression_rose(df,'ELIMINATION-1')
```

```python
df['First Impression Rose?'].value_counts()
```

```python
df.loc[:, 'First Impression Rose?'] = df['First Impression Rose?'].fillna(False)
```

```python
df['First Impression Rose?'].value_counts()
```

```python
## dropping header rows i.e., rows that don't refer to contestants
df.drop(df.loc[df['CONTESTANT']=='ID'].index, inplace=True)
```

```python
df
```

```python
df['Roses from Dates'] = (df[['ELIMINATION-1','ELIMINATION-2','ELIMINATION-3','ELIMINATION-4','ELIMINATION-5']] == "R").sum(axis=1)
```

```python
df
```

```python
## Contestants with value "W" in week 10 are the winners. 
## Contestants who have not been eliminated by week 7 are in the top 6
```

```python
eliminated = ['E','EQ','EF','ED','EU']
```

```python
type(eliminated)
```

```python
def early_elim(df, eliminated):
    return df.iloc[:,[3,4,5,6,7,8]].isin(eliminated).any(axis=1)
```

```python
df['Eliminated Early'] = early_elim(df,eliminated)
```

```python
df
```

```python
# Top 6 is our target variable
df['Top 6'] = ~df['Eliminated Early']
```

```python
df
```

```python
corr_coef = df['Roses from Dates'].corr(df['Top 6'])

print(corr_coef)
```

```python
corr_coef = df['First Impression Rose?'].corr(df['Top 6'])

print(corr_coef)
```

```python
# now we want to see whether or not they got a One-on-One date in the first 5 weeks
date = ['D1']
```

```python
def early_oneonone(df, date):
    return df.iloc[:,[13,14,15,16,17]].isin(date).any(axis=1)
```

```python
df['Early One-on-One'] = early_oneonone(df,date)
```

```python
df['Top 6']
```

Now we have the target variable: Top 6

Independent variables/features: First Impression Rose, the total number of roses they got within the first 5 weeks, 
and whether or not they got a one-on-one date in the first 5 weeks.

We can drop the columns we aren't interested in/that won't help our model.

```python
df = df.drop(columns=['SHOW','SEASON','CONTESTANT','ELIMINATION-1','ELIMINATION-2','ELIMINATION-3',
                'ELIMINATION-4','ELIMINATION-5','ELIMINATION-6','ELIMINATION-7','ELIMINATION-8','ELIMINATION-9',
                'ELIMINATION-10','DATES-1','DATES-2','DATES-3','DATES-4','DATES-5','DATES-6','DATES-7',
                'DATES-8','DATES-9','DATES-10','Eliminated Early'])
```

Great! We have these functional features that tell us a lot about each contestant. But they can't be processed
by Python. We need to encode them numerically.

We'll use dummy encoding by converting it to int.

```python
df['FIR_encoded'] = df['First Impression Rose?'].astype(int)
df['Early_1on1_encoded'] = df['Early One-on-One'].astype(int)
df['Target'] = df['Top 6'].astype(int)
```

```python
df = df.drop(columns=['First Impression Rose?','Top 6','Early One-on-One'])
```

```python
df
```

Looking good! These are ready to be processed by our model. But first, we need to address the class label imbalance.

There are more contestants that did not make the top 6, than those who did. It may be worthwhile to undersample the contestants who were eliminated early.

```python
import pandas as pd
from sklearn.utils import resample

class_counts = df['Target'].value_counts()
print(class_counts)
```

```python
minority_class_size = class_counts.min()
majority_class_size = minority_class_size
undersampled_data = pd.concat([resample(df[df['Target'] == 0], 
                                        replace=False,
                                        n_samples=majority_class_size),
                               df[df['Target'] == 1]])
```

```python
undersampled_class_counts = undersampled_data['Target'].value_counts()
print(undersampled_class_counts)
```
